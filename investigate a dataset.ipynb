```python
def splitcolumn(data):
    return data.str[0:].str.split('|',expand = True)

Genresdf = splitcolumn(df['Genres'])
Castsdf = splitcolumn(df['Cast'])
Keywordsdf = splitcolumn(df['Keywords'])
Companies = splitcolumn(df['ProductionCompanies'])
genres = []
for x in Genresdf:
    for z in Genresdf[x].unique():
        if z not in genres and z is not np.nan and z is not None : genres.append(z)
            
genres = sorted(genres)
print(genres)
for genre in genres: df[genre]=df['Genres'].str.contains(genre)
    
for newdf in [Genresdf,Castsdf,Keywordsdf,Companies]:
    newdf['ID'] = df['ID']




```

```python
def meltthis(data ):
    return pd.melt(data,col_level = 0,id_vars = 'ID',value_vars = [0,1,2,3,4]) #keep id column apply merge on rest of the columns

meltedgenres=meltthis(Genresdf)
meltedcasts = meltthis(Castsdf)
meltedkeywords = meltthis(Keywordsdf)
meltedcompanies = meltthis(Companies)

melteddf = [meltedgenres,meltedcasts,meltedkeywords,meltedcompanies]
for mdf in melteddf:
    mdf.columns = ['ID','Item','Value']
    mdf.drop(mdf.columns[1],inplace=True,axis=1)
    mdf.dropna(inplace = True)
```

```python
def Select(Field,xlabelCaption="\r",ylabelCaption="",ChartTitle="", Top=10):
    df_sorted = pd.DataFrame(df[Field].sort_values(ascending=False))[:Top]
    df_sorted['OriginalTitle'] = df['OriginalTitle']
    plt.figure(figsize=(12,6))
    
    Mean = np.mean(df[Field])   
    sns.barplot(x=Field, y='OriginalTitle', data=df_sorted, label=Field)
    plt.axvline(Mean, color='white', ls='dashed',linewidth=2.0, label='Means')
    plt.xlabel("{} {}".format(Field.capitalize() , xlabelCaption))
    
    plt.ylabel(ylabelCaption)
    plt.title(ChartTitle)

Select('Profit',Top=5)
```
